## Time Series Analysis

```python
# Ensure datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract time features
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.to_period('Q')
df['Week'] = df['Order Date'].dt.isocalendar().week

# Lead time
df['Ship Delay (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
# 1 Monthly Revenue Trend
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()

plt.figure(figsize=(14,5))
plt.plot(monthly_sales)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# 2 Profit Trend by Year & Quarter
quarter_profit = df.groupby('Quarter')['Profit'].sum()

plt.figure(figsize=(14,5))
quarter_profit.plot(kind='bar')
plt.title('Quarterly Profit Trend')
plt.xlabel('Quarter')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()

```

### This preliminary analysis shows us that there is a rise in sales and profit every end-of-year(4th quartile), but further analysis is required to pin point the exact reason for this trend

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure datetime columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract Quarter from Order Date
df['Quarter'] = df['Order Date'].dt.to_period('Q')

# Grouping
cat_contrib = df.groupby(['Quarter', 'Category'])[['Sales', 'Profit']].sum()
cat_contrib['Margin_%'] = (cat_contrib['Profit'] / cat_contrib['Sales']) * 100

# Reset index for plotting
cat_contrib_plot = cat_contrib.reset_index()

# Convert Quarter to string so seaborn doesnt choke
cat_contrib_plot['Quarter'] = cat_contrib_plot['Quarter'].astype(str)

# Sort quarter order for proper x-axis sequence
cat_contrib_plot = cat_contrib_plot.sort_values(by=['Quarter','Category'])

sns.set(style="whitegrid")

# 1 Quarterly Sales Contribution by Category (Bar Plot)
plt.figure(figsize=(14,6))
sns.barplot(
    data=cat_contrib_plot,
    x='Quarter',
    y='Sales',
    hue='Category'
)
plt.title("Quarterly Sales Contribution by Category")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.legend(title="Category")
plt.tight_layout()
plt.show()

# 2 Quarterly Profit Margin by Category (Line Plot)
plt.figure(figsize=(14,6))
sns.lineplot(
    data=cat_contrib_plot,
    x='Quarter',
    y='Margin_%',
    hue='Category',
    marker='o',
    sort=False
)
plt.title("Quarterly Profit Margin by Category")
plt.xlabel("Quarter")
plt.ylabel("Profit Margin (%)")
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```

##### based on catergorical time analysis we see slight increases in profit margin of office supply contribution in 4th quartile's of the year(specifically in 2014 and 2016). But the sales increase is uniform across all the categories(no specific category dominates or is cause of increased sales)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Quarter
driver_q = df.groupby('Quarter')[['Quantity', 'Discount']].mean().reset_index()
driver_q['Quarter'] = driver_q['Quarter'].astype(str)

sns.set(style="whitegrid")

#  Plot Quantity (on its own)
plt.figure(figsize=(15,5))
sns.lineplot(
    data=driver_q,
    x='Quarter',
    y='Quantity',
    marker='o'
)
plt.title("Average Quantity per Order by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Quantity")
plt.grid(True)
plt.tight_layout()
plt.show()

#  Plot Discount (on its own)
plt.figure(figsize=(15,5))
sns.lineplot(
    data=driver_q,
    x='Quarter',
    y='Discount',
    marker='o'
)
plt.title("Average Discount by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Average Discount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### We see a rise in quantity(of individual units) for 4th quartiles for year 2015-16-17 and rise in average discount in Q4 of 2014 these could be a possible reason for rises in total sales/profits every 4th quartile

### Q4 sales growth is primarily seasonal and spread across product categories. In 2014, the growth was supported by higher discounting. In later years (201517), increased unit volumes rather than discounting drove the uplift. Margin improvements, when they appear, are partly helped by stronger performance of Office Supplies in specific years (2014 & 2016), but this effect is not consistent across the dataset.

### California and New york have the highest sales and profit, New york being more profitable than California

```python


def profit_color(x):
    return 'Loss' if x < 0 else 'Profit'

# ---- California ----
df_cal = df[df['State'] == 'California']
df_cal_customer = df_cal.groupby('Customer Name')[['Sales', 'Profit']].sum()
df_cal_customer['ProfitType'] = df_cal_customer['Profit'].apply(profit_color)

# ---- New York ----
df_new = df[df['State'] == 'New York']
df_new_customer = df_new.groupby('Customer Name')[['Sales', 'Profit']].sum()
df_new_customer['ProfitType'] = df_new_customer['Profit'].apply(profit_color)

# ---- Subplots ----
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

palette = {'Profit': 'black', 'Loss': 'red'}
hue_order = ['Profit', 'Loss']

# --- California ---
sns.scatterplot(
    data=df_cal_customer,
    x='Sales', y='Profit',
    hue='ProfitType',
    palette=palette,
    hue_order=hue_order,
    ax=axes[0]
)
axes[0].set_title("California: Customer Sales vs Profit")
axes[0].set_xlim(0, 14000)
axes[0].set_ylim(-1000, 5000)

# --- New York ---
sns.scatterplot(
    data=df_new_customer,
    x='Sales', y='Profit',
    hue='ProfitType',
    palette=palette,
    hue_order=hue_order,
    ax=axes[1]
)
axes[1].set_title("New York: Customer Sales vs Profit")
axes[1].set_xlim(0, 14000)
axes[1].set_ylim(-1000, 5000)

# Consistent legend
handles, labels = axes[0].get_legend_handles_labels()
axes[0].legend(handles, labels, title="Profit Type")
axes[1].legend(handles, labels, title="Profit Type")

plt.tight_layout()
plt.show()
```

### How concentrated are the sales and Profit across customers?

```python
# --- Prepare data ---
df_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=True)
cum_sales = df_sales.cumsum() / df_sales.sum()

df_profit = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=True)
cum_profit = df_profit.cumsum() / df_profit.sum()

# --- Create subplots ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# =========================
#  Plot 1  Sales Pareto
# =========================
axes[0].plot(cum_sales.values)
axes[0].axhline(0.6, color='red', linestyle='--')
axes[0].axvline(int(len(df_sales) * 0.7), color='blue', linestyle='--')

axes[0].set_title("Customer Pareto Effect  Sales", fontsize=14)
axes[0].set_xlabel("Customers (sorted)", fontsize=12)
axes[0].set_ylabel("Cumulative Share of Sales", fontsize=12)

# =========================
#  Plot 2  Profit Pareto
# =========================
axes[1].plot(cum_profit.values)
axes[1].axhline(0.6, color='red', linestyle='--')
axes[1].axvline(int(len(df_profit) * 0.7), color='blue', linestyle='--')

axes[1].set_title("Customer Pareto Effect  Profit", fontsize=14)
axes[1].set_xlabel("Customers (sorted)", fontsize=12)
axes[1].set_ylabel("Cumulative Share of Profit", fontsize=12)

plt.tight_layout()
plt.show()
```

### Curve rises sharply = strong Pareto effect.

```python
# --- SALES ---
df_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)

cutoff_sales = df_sales.quantile(0.70, interpolation='higher')
top30_sales_total = df_sales[df_sales >= cutoff_sales].sum()
sales_share = top30_sales_total / df_sales.sum()

print(f"Top 30% customers contribute {sales_share:.1%} of total SALES.")


# --- PROFIT ---
df_profit = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False)

cutoff_profit = df_profit.quantile(0.70, interpolation='higher')
top30_profit_total = df_profit[df_profit >= cutoff_profit].sum()
profit_share = top30_profit_total / df_profit.sum()

print(f"Top 30% customers contribute {profit_share:.1%} of total PROFIT.")
```

### Top 30% of customers drives the majority of revenue, highlighting the importance of prioritizing high-value customers.

```python

# Build customer-level dataset
df_customer = df.groupby('Customer Name')[['Sales', 'Profit']].sum()

# Create a proper categorical column (not 'red'/'black')
df_customer['ProfitType'] = df_customer['Profit'].apply(
    lambda x: 'Loss' if x < 0 else 'Profit'
)

# Palette and order (consistent colors)
palette = {'Profit': 'black', 'Loss': 'red'}
hue_order = ['Profit', 'Loss']

plt.figure(figsize=(10,7))

sns.scatterplot(
    data=df_customer,
    x='Sales',
    y='Profit',
    hue='ProfitType',
    palette=palette,
    hue_order=hue_order,
    s=60
)

plt.title("Customer Sales vs Profit (All Customers)", fontsize=16)
plt.xlabel("Sales", fontsize=14)
plt.ylabel("Profit", fontsize=14)
plt.legend(title="Profit Type")

plt.show()
```

### The highest-revenue customer is currently unprofitable, signaling a margin issue that needs deeper analysis (pricing, discount levels, or fulfillment costs).

## Final Conclusion  Key Takeaways from the Analysis

- Sales consistently rise in the last quarter (Q4) each year across all product categories.
- In recent years, this rise is mainly due to customers buying more items, not because of high discounts, which helps profit remain strong.
- A small group (top 30%) of customers generates most of the companys revenue and almost all of its profit.
- This creates high value but also high dependency on those customers, meaning losing a few could significantly impact the business.
- New York and California lead in total sales, but New York generates higher profit from those sales.

### Overall Conclusion
The company should:
- Focus on improving and maintaining profitability from its most important customers
- Continue building Q4 performance without relying too much on discounts
- Apply successful practices from strong regions like New York to improve weaker areas
